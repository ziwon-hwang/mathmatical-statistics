가설:H0:모기의 흡혈시간 모평균!=180초.

     H1:모기의 흡혈시간 모평균=180초
     
기초통계량:n=5,표본평균=284.2,표준편차= 89.58602

검정통계량과 H0하에서의 분포:T= ~T(4)

pvalue:0.059996

결론:pvalue=0.059996 >유의수준 0.05이므로 H0을 기각하지 않는다.

```python
df=pd.DataFrame([176.0,202.9,315.0,374.6,352.5])
```

```python
dd=df.describe().T
dd[['mean','std']]
```

```python
stats.ttest_1samp([176.0,202.9,315.0,374.6,352.5],180)
```

# 일표본 모비율 z검정

가설:H0: 모평균μ=0.456(예전과 같다.)

     H1: 모평균μ!=0.456(변화 있다.)
     '

검정통계량과 H0하에서의 분포:Z=

pvalue:0.4979(양측 검정시)
    
결론:pvalue=0.4979 >유의수준 0.05이므로 H0을 기각하지 않는다..

# 일표본 t검정

가설:H0: μ=518(학생들의 평균성적이 한국평균과 같다.)
     H1: μ!=518(학생들의 평균성적이 한국평균과 같지 않다.)
     
기초통계량:n=100,표본평균=504.766 ,표준편차= 63.271

검정통계량과 H0하에서의 분포:T=  ~T(99)

검정통계량값:T=(504.766-518)/(63.271/10)=-2.092
    
pvalue:0.039(양측 검정시)
    
결론:pvalue=0.039 < 유의수준 0.05이므로 H0을 기각한다.
    
학생들의 평균성적이 한국평균과 같지 않다.

신뢰구간:(494.3588, 515.1732)

```python
DF=pd.read_csv('toefl.csv')
DF['grade']=np.round(((DF['listen']+DF['grammar']+DF['reading'])/3)*10,1)
DF['grade'].describe()
```

```python
stats.ttest_1samp(DF['grade'],518)
```

```python
#신뢰구간
print(504.766-norm.ppf(0.95,0,1)*(63.271/sqrt(100)),
      504.766+norm.ppf(0.95,0,1)*(63.271/sqrt(100)))
```

# 일표본 모비율 z검정

가설:H0: 모평균μ<=0.4(여학생 비율이 40% 이하이다)
     H1: 모평균μ> 0.4(40%보다 높아졌다)
기초통계량:n=100,모비율평균= 0.59,모비율표준편차= 0.0492

검정통계량과 H0하에서의 분포:Z

검정통계량값:Z=(0.59-0.4)/(0.049)=3.878
    
pvalue:0.00004803(단측 검정시)
    
결론:pvalue=0.00004803 <유의수준 0.05으로 H0을 기각한다
여학생 비율이 40% 이상이다.


```python
stats.binom_test(x=59,n=100,p=0.4,alternative='greater')
```
